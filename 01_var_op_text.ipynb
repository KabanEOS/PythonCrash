```python
# string formatting
greeting = 'Hello'
name = 'Michael'
cheer = 'Take a beer'

same_message_1 = '{}, {}. Welcome! {}'.format(greeting, name, cheer)
same_message_2 = f'{greeting}, {name.upper()}. Welcome! {cheer}'
print(same_message_1)
print(same_message_2)
```
