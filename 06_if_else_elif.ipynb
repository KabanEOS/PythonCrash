```python
# zero and empty sequences/map are seen as false when if'ed
zero = 0
empty_list = []
empty_tuple = ()
empty_set = set()
empty_dict = {}

if zero:
    print('Evaluated to true')
else:
    print('Evaluated to false')
```

```python
if empty_list:
    print('Evaluated to true')
else:
    print('Evaluated to false')
```

```python
if empty_tuple:
    print('Evaluated to true')
else:
    print('Evaluated to false')
```

```python
if empty_set:
    print('Evaluated to true')
else:
    print('Evaluated to false')
```

```python
if empty_dict:
    print('Evaluated as false')
else:
    print('Evaluated as false')
```
