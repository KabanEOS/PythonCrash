```python
# An empty list
empty_list = ()
empty_list = list()
print (empty_list)
```

```python
# Adds List Element as value of List.
List = ['Mathematics', 'chemistry', 1997, 2000]
List.append(20544)
print(List)
```

```python
courses = ['History', 'Math', 'Sport', 'Programming']
courses_2 = ['CockRiding','CowMilking']
nums = [3,4,6,2,1,5,4,2]
#slicing

#last one with -1
print(courses[-1])

#left slice
print(courses[:2])
print(courses[2:])
```

```python
courseStrJoined =', '.join(courses)
print(courseStrJoined)
```

```python
#split list
courseStrSplitted = courseStrJoined.split(', ')
print(courseStrSplitted)
```

```python
#appending
courses.append('Art')
courses.append('Zumba')
```

```python
#inserting
courses.insert(0, 'Poetry')
```

```python
#extending
courses.extend(courses_2)
```

```python
#removing
courses.remove('Math')
```

```python
#pop - delete last element from list and print it
courses.pop()
```

```python
#popping  - save to variable
popped = courses.pop()
print(popped)
```

```python
#reversing
courses.reverse()
```

```python
#sort FUNCTION - does NOT AFFECT original list
sortedCourses = sorted(courses)
print(sortedCourses)
```

```python
#sort METHOD - just sort the list
nums.sort()
courses.sort()
```

```python
#min, max
print(min(nums))
print(min(courses))
print(max(nums))
print(max(courses))
```

```python
#count
List = [1, 2, 3, 1, 2, 1, 2, 3, 2, 1]
print(List.count(1))
```

```python
#length
print(len(List))
```

```python
#sum
print(sum(nums))
```

```python
#index
print(courses.index('Poetry'))
```

```python
#List.index(element, start, end)
List = [1, 2, 3, 1, 2, 1, 2, 3, 2, 1]
print(List.index(2))
print(List.index(2,2))
print(List.index(2,5,7))
```

```python
# if exist method - return boolean
print('Math' in courses)
print('Poetry' in courses)
```

```python
#for loop to list list
for course in courses:
    print(course)
```

```python
#enumerated for lopp
for index, course in enumerate(courses, start=1):
    print(index, course )
```
