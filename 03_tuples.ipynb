```python
# An empty tuple
empty_tuple = ()
empty_tuple = tuple()
print (empty_tuple)
```

```python
#MUTABLE operation is possible with lists
list_1 = ['History', 'Math', 'Sport', 'Programming']
list_2 = list_1

print(list_1)
print(list_2)

list_1[0] = 'Art'

print(list_2)
print(list_1)

#object has changed in both original and modified place
```

```python
# IMMUTABLE
tuple_1 = ('History', 'Math', 'Sport', 'Programming')
tuple_2 = tuple_1

print(tuple_1)
print(tuple_2)

# tuple_1[0] = 'Art'#TypeError: 'tuple' object does not support item assignment

print(tuple_1)
print(tuple_2)
```

```python
# Creating non-empty tuples

# One way of creation
tup = 'python', 'geeks'
print(tup)

# Another for doing the same
tup = ('python', 'geeks')
print(tup)
```

```python
# Nested tuples
tuple1 = (0, 1, 2, 3)
tuple2 = ('python', 'geek')
tuple3 = (tuple1, tuple2)
print(tuple3)

```
